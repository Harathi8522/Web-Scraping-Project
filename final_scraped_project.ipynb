```
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```
base_url="https://books.toscrape.com/catalogue/page-{}.html"
books=[]
#Loop through all 50 pages
for page in range(1,51):
  url=base_url.format(page)
  response=requests.get(url)
  if response.status_code!=200:
    print(f"Failed to fetch page {page}. Status code: {response.status_code}")
    continue
  soup=BeautifulSoup(response.text,'html.parser')
  for book in soup.find_all('article',class_='product_pod'):
    title=book.h3.a['title']
    price=book.find('p',class_='price_color').text
    # Extract rating class, handling potential None if class is missing
    rating_element = book.p.get('class')
    rating = rating_element[1] if rating_element and len(rating_element) > 1 else "N/A"
    link=book.h3.a['href']
    books.append({'Title':title,
                  'Price':price,
                  'Rating':rating,
                  'Link':f"https://books.toscrape.com/catalogue/{link}"
                  })
    print(f"Total books scraped:{len(books)}") # Moved inside the inner loop to show progress

df = pd.DataFrame(books) # Moved outside the loop to create DataFrame once
df.to_csv('all_books.csv', index=False) # Moved outside the loop to save once
print("Data saved to all_books.csv")
```
